```python
# Görüntüyü yükle
image = cv2.imread("dataset/img2.jpg")

# Gri tonlamaya çevir
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Gürültü azaltma (Gaussian Blur)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Uyarlanabilir threshold
adaptive_thresh = cv2.adaptiveThreshold(
    blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
)

# Morfolojik işlemler (Kenarlık yumuşatma)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
processed = cv2.morphologyEx(adaptive_thresh, cv2.MORPH_CLOSE, kernel)

plt.imshow(processed, cmap='gray')
plt.title("Processed Image")
plt.show()

#########################################

# Açılma işlemi
kernel = np.ones((3, 3), np.uint8)
opened_image = cv2.morphologyEx(processed, cv2.MORPH_OPEN, kernel)

# Medyan filtresi uygula
cleaned_image = cv2.medianBlur(opened_image, 3)

plt.imshow(cleaned_image, cmap='gray')
plt.title("Cleaned Image")
plt.show()

#############################

# Erozyon işlemi
erosion_kernel = np.ones((2, 2), np.uint8)
eroded_image1 = cv2.erode(cleaned_image, erosion_kernel, iterations=1)

# Sonucu göster
plt.imshow(eroded_image1, cmap='gray')
plt.title("Eroded Image")
plt.show()

text = pytesseract.image_to_string(eroded_image1, lang='tur')
print(text)
```

# TKINTER ile Arayüz Oluşturulması

```python
import cv2
import numpy as np
import pytesseract
import tkinter as tk
from tkinter import filedialog, Button, Frame, Text, Scrollbar, RIGHT, Y
from PIL import Image, ImageTk

def choose_image():
    global img_display, processed_display, img_path, original_image, processed_image
    img_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png;*.bmp")])
    if img_path:
        original_image = cv2.imread(img_path)
        image = Image.open(img_path)
        
        # Görüntüyü maksimum boyutlara sığacak şekilde yeniden boyutlandır (250x250), oranı koruyarak
        image.thumbnail((250, 250))
        img_display = ImageTk.PhotoImage(image)

        # Kanvası, görüntü ile güncelle ve kanvas boyutunu resmin boyutuna uyacak şekilde ayarla
        img_canvas.config(width=image.width, height=image.height)
        img_canvas.create_image(0, 0, anchor=tk.NW, image=img_display)
        img_canvas.config(scrollregion=img_canvas.bbox(tk.ALL))

        # İşlenmiş görüntü alanını temizle
        processed_canvas.delete("all")

def convert_to_text():
    global processed_display, processed_image
    if img_path:
        # Orijinal resmi oku
        gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
        # GaussianBlur uygula
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)
        # Uyarlanabilir eşikleme
        adaptive_thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
        # Morfolojik işlemler (Kenarları yumuşatma)
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
        processed = cv2.morphologyEx(adaptive_thresh, cv2.MORPH_CLOSE, kernel)

        kernel = np.ones((2, 2), np.uint8)
        opened_image = cv2.morphologyEx(processed, cv2.MORPH_OPEN, kernel)

        # Medyan filtresi uygula
        cleaned_image = cv2.medianBlur(opened_image, 3)

        # Erozyon işlemi
        erosion_kernel = np.ones((2, 2), np.uint8)
        processed_image = cv2.erode(cleaned_image, erosion_kernel, iterations=1)

        # İşlenmiş resmi görüntü için dönüştür
        processed_pil = Image.fromarray(cv2.cvtColor(processed_image, cv2.COLOR_GRAY2RGB))
        processed_pil.thumbnail((250, 250))
        processed_display = ImageTk.PhotoImage(processed_pil)
        
        # İşlenmiş kanvası, görüntü ile güncelle ve kanvas boyutunu resmin boyutuna uyacak şekilde ayarla
        processed_canvas.config(width=processed_pil.width, height=processed_pil.height)
        processed_canvas.create_image(0, 0, anchor=tk.NW, image=processed_display)
        processed_canvas.config(scrollregion=processed_canvas.bbox(tk.ALL))

        # OCR ile metin çıkarma
        text = pytesseract.image_to_string(processed_image, lang='tur')

        # Metni text widget'ında göster
        text_widget.delete(1.0, tk.END)
        text_widget.insert(tk.END, text)

# Tkinter penceresini başlat
root = tk.Tk()
root.title("Resimden Metne OCR")

# Ana çerçeve oluştur
main_frame = Frame(root, width=800, height=400)
main_frame.pack(padx=10, pady=10)

# Orijinal resim gösterim alanı (şimdi Kanvas kullanılıyor)
img_canvas = tk.Canvas(main_frame, bg="gray")
img_canvas.grid(row=0, column=0, padx=10, pady=10)

# İşlenmiş resim gösterim alanı (şimdi Kanvas kullanılıyor)
processed_canvas = tk.Canvas(main_frame, bg="gray")
processed_canvas.grid(row=0, column=1, padx=10, pady=10)

# Butonlar için alan
button_frame = Frame(main_frame)
button_frame.grid(row=1, column=0, columnspan=2, pady=10)

choose_btn = Button(button_frame, text="Resim Seç", command=choose_image)
choose_btn.grid(row=0, column=0, padx=5)

convert_btn = Button(button_frame, text="Metne Dönüştür", command=convert_to_text)
convert_btn.grid(row=0, column=1, padx=5)

# Metin çıktısı alanı
text_output_frame = Frame(root)
text_output_frame.pack(padx=10, pady=10, fill="both", expand=True)

scrollbar = Scrollbar(text_output_frame)
scrollbar.pack(side=RIGHT, fill=Y)

text_widget = Text(text_output_frame, wrap="word", yscrollcommand=scrollbar.set, height=10, width=60)
text_widget.pack(padx=10, pady=10, fill="both", expand=True)
scrollbar.config(command=text_widget.yview)

root.mainloop()
```

```python
import cv2
import numpy as np
import pytesseract
import tkinter as tk
from tkinter import filedialog, Button, Frame, Text, Scrollbar, RIGHT, Y
from PIL import Image, ImageTk
import pandas as pd

# Gıda maddelerini CSV'den al
def load_food_data():
    global food_list, food_data
    food_data = pd.read_csv("ambalajli_gida_urunleri.csv")  # Dosya yolunu güncelleyin
    food_list = food_data['Maddeler'].tolist()

def choose_image():
    global img_display, processed_display, img_path, original_image, processed_image
    img_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png;*.bmp")])
    if img_path:
        original_image = cv2.imread(img_path)
        image = Image.open(img_path)
        
        # Görüntüyü maksimum boyutlara sığacak şekilde yeniden boyutlandır (250x250), oranı koruyarak
        image.thumbnail((250, 250))
        img_display = ImageTk.PhotoImage(image)  # Resim nesnesini global tutuyoruz

        # Kanvası, görüntü ile güncelle ve kanvas boyutunu resmin boyutuna uyacak şekilde ayarla
        img_canvas.config(width=image.width, height=image.height)
        img_canvas.create_image(0, 0, anchor=tk.NW, image=img_display)
        img_canvas.config(scrollregion=img_canvas.bbox(tk.ALL))

        # İşlenmiş görüntü alanını temizle
        processed_canvas.delete("all")

def convert_to_text():
    global processed_display, processed_image
    if img_path:
        # Orijinal resmi oku
        gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
        # GaussianBlur uygula
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)
        # Uyarlanabilir eşikleme
        adaptive_thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
        # Morfolojik işlemler (Kenarları yumuşatma)
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
        processed = cv2.morphologyEx(adaptive_thresh, cv2.MORPH_CLOSE, kernel)

        kernel = np.ones((2, 2), np.uint8)
        opened_image = cv2.morphologyEx(processed, cv2.MORPH_OPEN, kernel)

        # Medyan filtresi uygula
        cleaned_image = cv2.medianBlur(opened_image, 3)

        # Erozyon işlemi
        erosion_kernel = np.ones((2, 2), np.uint8)
        processed_image = cv2.erode(cleaned_image, erosion_kernel, iterations=1)

        # İşlenmiş resmi görüntü için dönüştür
        processed_pil = Image.fromarray(cv2.cvtColor(processed_image, cv2.COLOR_GRAY2RGB))
        processed_pil.thumbnail((250, 250))
        processed_display = ImageTk.PhotoImage(processed_pil)  # İşlenmiş resmi global tutuyoruz
        
        # İşlenmiş kanvası, görüntü ile güncelle ve kanvas boyutunu resmin boyutuna uyacak şekilde ayarla
        processed_canvas.config(width=processed_pil.width, height=processed_pil.height)
        processed_canvas.create_image(0, 0, anchor=tk.NW, image=processed_display)
        processed_canvas.config(scrollregion=processed_canvas.bbox(tk.ALL))

        # OCR ile metin çıkarma
        text = pytesseract.image_to_string(processed_image, lang='tur')

        # Metni text widget'ında göster
        text_widget.delete(1.0, tk.END)
        text_widget.insert(tk.END, text)

        # Çıkan metinde maddeleri arayıp yazdır
        check_for_food_items(text)

def check_for_food_items(text):
    found_items = []
    # CSV'deki gıda maddeleri ve yanındaki bilgileri kontrol et
    for index, row in food_data.iterrows():
        food_item = row['Maddeler']
        if food_item.lower() in text.lower():
            found_items.append((food_item, row['Zararlılık']))  # Gıda maddesini ve zararlılık bilgisini sakla
    
    # Eğer madde bulunmuşsa, text kısmında listele
    if found_items:
        text_widget.insert(tk.END, "\n\nBulunan Gıda Maddeleri ve Bilgileri:\n")
        for item, info in found_items:
            text_widget.insert(tk.END, f"- {item}: {info}\n")
    else:
        text_widget.insert(tk.END, "\n\nBulunan gıda maddesi yok.\n")

# Tkinter penceresini başlat
root = tk.Tk()
root.title("Resimden Metne OCR")

# Gıda maddelerini yükle
load_food_data()

# Ana çerçeve oluştur
main_frame = Frame(root, width=800, height=400)
main_frame.pack(padx=10, pady=10)

# Orijinal resim gösterim alanı (şimdi Kanvas kullanılıyor)
img_canvas = tk.Canvas(main_frame, bg="gray")
img_canvas.grid(row=0, column=0, padx=10, pady=10)

# İşlenmiş resim gösterim alanı (şimdi Kanvas kullanılıyor)
processed_canvas = tk.Canvas(main_frame, bg="gray")
processed_canvas.grid(row=0, column=1, padx=10, pady=10)

# Butonlar için alan
button_frame = Frame(main_frame)
button_frame.grid(row=1, column=0, columnspan=2, pady=10)

choose_btn = Button(button_frame, text="Resim Seç", command=choose_image)
choose_btn.grid(row=0, column=0, padx=5)

convert_btn = Button(button_frame, text="Metne Dönüştür", command=convert_to_text)
convert_btn.grid(row=0, column=1, padx=5)

# Metin çıktısı alanı
text_output_frame = Frame(root)
text_output_frame.pack(padx=10, pady=10, fill="both", expand=True)

scrollbar = Scrollbar(text_output_frame)
scrollbar.pack(side=RIGHT, fill=Y)

text_widget = Text(text_output_frame, wrap="word", yscrollcommand=scrollbar.set, height=10, width=60)
text_widget.pack(padx=10, pady=10, fill="both", expand=True)
scrollbar.config(command=text_widget.yview)

root.mainloop()
```
